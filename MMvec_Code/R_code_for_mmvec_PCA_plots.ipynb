```r
###The "pca3d" package was used to create our mmvec PCs.
install.packages("pca3d")
library("pca3d")
```

```r
###Loading in our conditionals file
cond <- read.csv("Conditionals_CoreMetabs_wNames_and_CF.csv")
```

```r
head(cond)
dim(cond)
```

```r
###Now turning that conditionals file into a PCA.
cond.pca <- prcomp(cond[,-1:-8],scale.=TRUE)
```

```r
###Saving our ClassyFire subclass label (column 8) as a grouping factor for coloring the plot. 
cond.gr <- factor(cond[,8])
```

```r
###Creating a 2D PCA using parameters from the package. This results in a biplot with two PCs and the three most influential taxa for each PC (6 total).
pca2d(cond.pca, components=1:3, biplot=TRUE, biplot.var=2, group=cond.gr, axes.color="black", radius=1.5)
```

```r
###Creating a 3D PCA with the same parameters, so there are 3 PCs each showing heir two most influential taxa (6 total).
pca3d(cond.pca, biplot=TRUE, biplot.var=2, group=cond.gr, axes.color="black",show.plane=FALSE, radius=2, legend="bottomright")
```

```r
###Saving the 3D PCA plot
snapshotPCA3d(file="PCA_3D.png")
```
