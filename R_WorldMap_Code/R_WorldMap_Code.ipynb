```r
###Combining these into a single 'colors' to be used by rworldmap
colors <- c("#32C4D6","#ABEE85","#FFFFCC")
```

```r
###First, create a dataframe with the countries you want to highlight. 
##Since we want Burkina Faso, Peru, and United States, we need to put 
###those in the data frame. There's multiple ways of specifying the 
###countries, but we'll use ISO3 codes.
popDF <- data.frame(country=c("BFA","PER","USA"),country=c(1,2,3))
```

```r
###Now with the data frame made, we need to tell R to create a map 
###from that list. Use the joinCountryData2Map function from rworldmap. 
###Specify the dataframe with the list of countries, how we specified 
###their names, and what regions we are using (countries). The code 
###will provide a quick output so you can see if it worked -- our 3 
###codes worked and are placed onto the map.
popMap <- joinCountryData2Map(popDF,joinCode="ISO3",nameJoinColumn = "country")
```

```r
###It's time to create the map. First, specify the file name and 
###dimensions for saving the image.
pdf("Rworldmap2.pdf",width=10,height=10)
```

```r
###Now we make the map. Use the mapCountryData function from rworldmap. 
###Input the map we want to make (popMap), the region/territory 
###(country), the category (categorical), the line width 
###(we use 1, but increasing this will make the borders thicker), the 
###color of the borders (default is gray, but we use black to make them 
###distinct), the colors for countries we don't want (colored gray), 
###and the colors for our countries (callback to the colors we 
###specified earlier). Last, use dev.off() to save the file.
mapCountryData(popMap,nameColumnToPlot = "country",catMethod="categorical",lwd=1,missingCountryCol = gray(0.8),colourPalette=colors)
```
